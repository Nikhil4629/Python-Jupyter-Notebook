```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months=np.array(['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec'])
sales=np.array([200,220,250,275,300,320,310,295,280,270,260,250])
```

```python
plt.figure(figsize=(6,4))
plt.plot(months,sales,marker='o',linestyle='-',color='b')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
 np.random.seed(0)
data=np.random.normal(loc=0,scale=1,size=1000)
plt.figure(figsize=(6,4))
plt.hist(data,bins=30,color='purple',edgecolor='black')
plt.title('Random Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
np.random.seed(1)
x=np.random.rand(50)*100
y=np.random.rand(50)*100
plt.figure(figsize=(6,4))
plt.scatter(x,y,color='red',alpha=0.7)
plt.title('Random Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```

```python
categories=np.array(['A','B','C','D','E'])
values=np.array([10,20,15,25,30])
plt.figure(figsize=(6,4))
plt.bar(categories,values,color=['blue','green','red','purple','orange'])
plt.title('Category values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(False)
plt.show()
```

```python
np.random.seed(2)
group1=np.random.normal(loc=0,scale=1,size=100)
group2=np.random.normal(loc=1,scale=1.5,size=100)
plt.figure(figsize=(6,4))
plt.boxplot([group1,group2],label=['Group 1','Group 2'])
plt.title('Box plot of two groups')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
np.random.seed(3)
data_matrix=np.random.rand(10,10)
plt.figure(figsize=(6,4))
plt.imshow(data_matrix,cmap='coolwarm',interpolation='none')
plt.colorbar(label='Value')
plt.title('Heatmap of random data matrix')
plt.show()
```

```python
x=np.linspace(0, 10, 100)
y1=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(6,2))
plt.subplot(1,2,1)
plt.plot(x,y1,'b-')
plt.title('Sine wave')
plt.subplot(1,2,2)
plt.plot(x,y2,'r--')
plt.title('Cosine wave')
plt.tight_layout()
plt.show()
```

```python
x=np.linspace(0, 10, 100)
y1=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(6,2))
plt.subplot(2,1,1)
plt.plot(x,y1,'b-')
plt.title('Sine wave')
plt.subplot(2,1,2)
plt.plot(x,y2,'r--')
plt.title('Cosine wave')
plt.tight_layout()
plt.show()
```
